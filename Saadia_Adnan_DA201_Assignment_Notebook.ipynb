### Data Cleaning

```python
# Converting date columns
appointments['appointment_month'] = pd.to_datetime(appointments['appointment_month'])
duration['appointment_date'] = pd.to_datetime(duration['appointment_date'])

# Standardise column names (lowercase)
appointments.columns = appointments.columns.str.lower()
duration.columns = duration.columns.str.lower()
categories.columns = categories.columns.str.lower()
```

### Demand Analysis - Total Appointments Over Time 

```python
# Grouping by month
monthly_demand = appointments.groupby('appointment_month')['count_of_appointments'].sum()

# Plotting
plt.figure(figsize=(10,5))
monthly_demand.plot()
plt.title('Total Appointments Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Appointments')
plt.show()
```

### Waiting Time Analysis

```python
# Waiting time distribution

import pandas as pd
import matplotlib.pyplot as plt

# Group data
waiting_time = appointments.groupby(
    'time_between_book_and_appointment'
)['count_of_appointments'].sum()

# Sort
waiting_time = waiting_time.sort_values()

# Plot
plt.figure(figsize=(10,6))
waiting_time.plot(kind='barh')

plt.title('Waiting Time Distribution')
plt.xlabel('Appointments')
plt.ylabel('Waiting Time')

plt.tight_layout()
plt.show()

```

## Appointmnet Mode Analysis

```python
# Appointment mode distribution
mode_analysis = appointments.groupby('appointment_mode')['count_of_appointments'].sum()

# Plot
plt.figure(figsize=(8,5))
mode_analysis.plot(kind='bar')
plt.title('Appointment Mode Distribution')
plt.ylabel('Number of Appointments')
plt.xticks(rotation=45)
plt.show()
```

## Appointment Status - Efficiency

```python
# Appointment status distribution
status_analysis = appointments.groupby('appointment_status')['count_of_appointments'].sum()

# Plotting a pie chart
plt.figure(figsize=(6,6))
status_analysis.plot(kind='pie', autopct='%1.1f%%')
plt.title('Appointment Status Distribution')
plt.ylabel('')
plt.show()
```

## Duration Analysis 

```python
# Duration distribution
duration_analysis = duration.groupby('actual_duration')['count_of_appointments'].sum()

# Sorting values
duration_analysis = duration_analysis.sort_values()

# Plotting bars 
plt.figure(figsize=(10,6))
duration_analysis.plot(kind='barh')
plt.title('Consultation Duration Distribution')
plt.xlabel('Appointments')
plt.ylabel('Duration')
plt.show()
```

## National Categories Analysis - Top 10

```python
# Top categories
category_analysis = categories.groupby('national_category')['count_of_appointments'].sum()

top_categories = category_analysis.sort_values(ascending=False).head(10)

# Plotting a bar chart 
plt.figure(figsize=(10,5))
top_categories.plot(kind='bar')
plt.title('Top 10 Appointment Categories')
plt.ylabel('Appointments')
plt.xticks(rotation=45)
plt.show()
```

## GP vs Other Staff Analysis 

```python
# Comparing the workload
hcp_analysis = appointments.groupby('hcp_type')['count_of_appointments'].sum()

# Plot

plt.figure(figsize=(6,4))
hcp_analysis.plot(kind='bar')
plt.title('Appointments by Healthcare Professional')
plt.ylabel('Appointments')
plt.xticks(rotation=45)
plt.show()
```

## Same Day vs Delayed Appointments

```python
# Creating simple grouping
appointments['wait_group'] = appointments['time_between_book_and_appointment'].apply(
    lambda x: 'Same Day' if 'Same Day' in str(x) else 'Delayed'
)

wait_analysis = appointments.groupby('wait_group')['count_of_appointments'].sum()

wait_analysis.plot(kind='bar', figsize=(6,4))
plt.title('Same Day vs Delayed Appointments')
plt.ylabel('Appointments')
plt.show()
```

**Question 3:** Which month had the highest number of appointments?

```python
# Number of appointments per month == sum of count_of_appointments by month.
# Use the groupby() and sort_values() functions.
# Determine the total number of appointments per month in the national_categories (nc) DataFrame.
# The groupby() function is used with year and month extracted from the appointment_date column, and the sum() function calculates the total number of appointments. 
# The results are sorted in descending order to identify the month with the highest number of appointments.

# Ensure datetime format
nc['appointment_date'] = pd.to_datetime(nc['appointment_date'])

# Group by year and month, then sum appointments
monthly_appointments = nc.groupby(
    [nc['appointment_date'].dt.year, nc['appointment_date'].dt.month]
)['count_of_appointments'].sum()

# Sort results (highest first)
monthly_appointments_sorted = monthly_appointments.sort_values(ascending=False)

# View output
monthly_appointments_sorted
```

### Analysis of Retweets and Favourite Counts

It would not be particularly useful to only focus on retweeted and favourite tweet messages when analysing healthcare-related Twitter data.

1: Highly Skewed Data 

The engagement data shows that the majority of tweets have very low or zero retweets and favourites. This means that only a small number of tweets gain significant visibility, while most tweets contribute little engagement.

2: Not Representative of Overall Trends

Focusing only on highly retweeted or favourited tweets would ignore the majority of the dataset. As a result, the analysis would not reflect the full range of healthcare discussions occurring on Twitter.

3: Bias Towards Popular or Influential Users

Tweets with high engagement are often posted by organisations, influencers, or accounts with large followings. This introduces bias, as these tweets may reflect promotional or institutional messages rather than general public opinion.

4: Limited Insight into Topic Trends

Retweets and favourites measure popularity rather than topic frequency. They do not provide direct insight into what topics are being discussed most frequently.

While engagement metrics can be useful in identifying influential or viral content, they are not suitable as the primary method for analysing healthcare trends. Instead, analysing hashtags and tweet content provides a more comprehensive and representative understanding of public discussions.

```python
# Create a new DataFrame with only the tweet text
tweets_text = tweets[['tweet_full_text']]

# View the new DataFrame
tweets_text.head()
```

A new DataFrame was created containing only the `tweet_full_text` column.

- The focus of this analysis is on tweet content and hashtags

```python
# Loop through the messages, and create a list of values containing the # symbol.
# Create an empty list to store hashtags
hashtags = []

# Loop through tweet messages
for tweet in tweets_text['tweet_full_text']:
    
    # Split each tweet into words
    words = tweet.split(' ')
    
    # Check each word
    for word in words:
        
        # If the word contains '#', add it to the list
        if '#' in word:
            hashtags.append(word.lower())

# Preview first 20 hashtags
hashtags[:20]
```
